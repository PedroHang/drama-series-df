```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('drama_series.csv')
```

```python
df = df.drop('Gross Revenue', axis=1)
```

```python
df.dropna(subset=['Runtime'], inplace=True)
```

```python
df['Runtime'] = df['Runtime'].str.replace('min', '')
```

```python
df['Runtime'] = df['Runtime'].str.replace(',', '.').str.replace('[^0-9.]', '', regex=True).astype(float)
```

```python
df
```

```python
g = sns.jointplot(x=df['Runtime'], y=df['Rating'], data=df,
                  kind="reg", truncate=False,
                  xlim=(0, 60), ylim=(0, 12),
                  color="m", height=7)
```
